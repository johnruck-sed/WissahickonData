```python
#Unaveraged Main Transect

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


df = np.asarray(pd.read_csv('Data\GrainSize - data.csv',header=1))
gvals = df[:,::2]
vvals = df[:,1::2]


plt.plot(gvals,vvals)
plt.title('Unaveraged Main Transect')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```
