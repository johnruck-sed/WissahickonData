```python
#un-averaged

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


df = np.asarray(pd.read_csv('Downloads\GrainSize - data.csv',header=1))
gvals = df[:,::2]
vvals = df[:,1::2]


plt.plot(gvals,vvals)
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```

```python
#averaged

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


df = np.asarray(pd.read_csv('Downloads\GrainSizeAverage.csv',header=1))
gvals = df[:,::2]
vvals = df[:,1::2]


plt.plot(gvals,vvals)
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Average Diff. % Volume(%)')
plt.show()
```
